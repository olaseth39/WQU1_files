```python
def describe(data, key):
    
    sorted_ = [script[key] for script in data]
    sorted_.sort()
    mid = len(sorted_)// 2
    
    total = sum(sorted_)
    avg = total / len(data)
    s = (sum([(script[key] - avg)**2 for script in data]) / len(data)) ** 0.5
    med = sorted_[len(data)// 2]
    
    q25 = sorted_[:mid][mid // 2]
    q75 = sorted_[mid:][mid // 2]
    
    return (total, avg, s, q25, med, q75)
    #return (total, avg, s, sorted_)
    
```

```python
def group_by_field_v1(data, fields):
    """
    Return a dictionary that has grouped data by all unique fields
    
    data: list of dictionaries
    fields: a tuple of the keys to be used for the grouping
    """
    group = {}
    
    for d in data:
        key = tuple(d[f] for f in fields)
        
        if key in group:
            group[key].append(d)
        else:
            group[key] = [d]
            
    return group
```

```python
def group_by_field_v2(data, fields):
    """
    Return a dictionary that has grouped data by all unique fields
    
    data: list of dictionaries
    fields: a tuple of the keys to be used for the grouping
    """
    unique_fields = {tuple(d[f] for f in fields) for d in data}
    groups = {name : [] for name in unique_fields}
    
    
    for d in data:
        key = tuple(d[f] for f in fields)
        groups[key].append(d)
            
    return groups
```

```python
group_by_field_v2(scripts, fields)
```

```python
def group_by_field_v3(data, fields):
    """
    Return a dictionary that has grouped data by all unique fields
    
    data: list of dictionaries
    fields: a tuple of the keys to be used for the grouping
    """
    groups = {}
    
    for d in data:
        key = tuple(d[f] for f in fields)
        groups[key] = groups.get(key, []) + [d]
            
    return groups
```

```python
from collections import defaultdict

def group_by_field_v4(data, fields):
    """
    Return a dictionary that has grouped data by all unique fields
    
    data: list of dictionaries
    fields: a tuple of the keys to be used for the grouping
    """
    groups = defaultdict(list)
    
    for d in data:
        key = tuple(d[f] for f in fields)
        groups[key].append(d)
            
    return groups
```
